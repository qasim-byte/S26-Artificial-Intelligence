```python
#lets say : 4214 6182 4263 9790 is card Number
card = input("Enter your credit card number: ")
card = card.replace(" ","") #Removing Spaces between Numbers
# require at least 16 digits and only digits
if len(card) >= 16 and card.isdigit():
    # keep original card for calculation, create a masked version for display
    check_digit = card[-1]
    masked_card = card[:-1] + "X"
    print(masked_card)
    # reverse the original card number string
    reversed_card = card[::-1]
    print("Reversed card number:", reversed_card)
    # now process digits excluding the checksum (which is at index 0 of reversed_card)
    doubled_digits = []
    for i in range(1, len(reversed_card)):  # start at 1 to exclude checksum
        digit = int(reversed_card[i])
        # in reversed order, positions 1,3,5,... are doubled
        if i % 2 == 1:
            doubled_digit = digit * 2
            if doubled_digit > 9:
                doubled_digit -= 9
            doubled_digits.append(doubled_digit)
        else:
            doubled_digits.append(digit)
    print("Doubled digits:", doubled_digits)
    # adding all the processed digits and also include the checksum digit
    total_sum = sum(doubled_digits) + int(reversed_card[0])
    print("Total sum:", total_sum)
    # checking if the total sum is divisible by 10
    if total_sum % 10 == 0:
        print("The credit card number is valid.")
    else:
        print("The credit card number is invalid.")
else:
    print("Invalid input. Please enter a valid credit card number with at least 16 digits.")
```
