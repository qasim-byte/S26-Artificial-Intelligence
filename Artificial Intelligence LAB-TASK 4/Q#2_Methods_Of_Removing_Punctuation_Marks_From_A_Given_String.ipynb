```python
import re       #Regular Expressions or Regex
import string
text = "My,N@me !$; Q@$;m"
p = string.punctuation
text1 = re.sub(f"[{re.escape(p)}]", "", text)
print(text1)
```

```python
punctuation = "!,@#$%^&*()-_=+{}[]:;\"'|\?/"
string = "My,N@me !$; Q@$;m"
list = []
for letter in string:
    if letter not in punctuation:
        list.append(letter)

final = "".join(list)
print(final)
```

```python
string = "My,N@me !$; Q@$;m"
for i in string:
    if i.isalnum() or i.isspace():
        print(i,end = "")
```
