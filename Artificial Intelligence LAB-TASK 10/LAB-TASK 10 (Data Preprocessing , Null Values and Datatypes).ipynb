<div 
  <h2 style='color: #58c2f7; font-family: Serif, sans-serif; font-size: 1.8em;'>Categorical Feature Analysis</h2>
 
</div>

```python
column = ['Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass']

for col in column:
    print(f'\n{col}:')
    v_count = data[col].value_counts().sort_index()
    for val, count in v_count.items():
        print(f"  {val} → {count} students ({round((count/len(data)*100), 2)}%)")
```
